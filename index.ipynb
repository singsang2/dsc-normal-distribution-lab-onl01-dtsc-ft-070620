## A quick refresher! 

Here's the formula for the normal distribution density function once more:

$$ \large N(x) = \dfrac{1}{\sigma \sqrt {2\pi }}e^{-\dfrac{(x-\mu)^2}{2\sigma^2}}$$

Here, 
- $\mu$ is the mean
- $\sigma$ is the standard deviation
- $\pi \approx 3.14159 $ 
- $ e \approx 2.71828 $


## First generate a normal distribution containing 5000 values with $\mu=14$ and $\sigma = 2.8$

```python
# Generate a random normal variable with given parameters , n=5000
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = np.random.normal(14, 2.8, 5000)
mu = 14
sig = 2.8
```

```python
fig, ax = plt.subplots(figsize=(12,8))
count, bins, ignored = ax.hist(n, bins=20, density=True)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.distplot(n, bins=20)
```

## Use the formula to calculate the density function with $\mu$, $\sigma$ and bin information obtained before

$$ \large N(x) = \dfrac{1}{\sigma \sqrt {2\pi }}e^{-\dfrac{(x-\mu)^2}{2\sigma^2}}$$

Here, 
- $\mu$ is the mean
- $\sigma$ is the standard deviation
- $\pi \approx 3.14159 $ 
- $ e \approx 2.71828 $


```python
# Calculate the normal Density function 
pi = np.pi
e = np.exp(1)

density = 1/(sig*np.sqrt(2*pi))*np.exp(-(n-mu)**2/(2*sig**2))
```

## Plot the histogram and density function

```python
# Plot histogram along with the density function
fig, ax = plt.subplots(figsize=(12,8))
count, bins, ignored = ax.hist(n, bins=20, density=True, alpha=0.8, label='PDF')
ax.scatter(n, density, color='orange', marker='.', s=1,  alpha=0.8, label='density graph')
ax.set(xlabel='value', ylabel='fraction', title='normal distribution')
```
